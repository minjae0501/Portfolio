## 전체 주유소 데이터 가져오기

```python
driver.find_element_by_xpath('''//*[@id="SIDO_NM0"]''').click() #지역 클릭
driver.find_element_by_xpath('''//*[@id="SIDO_NM0"]/option[2]''').click() # 서울 클릭
time.sleep(2)
```

```python
si_gun_gu_raw = driver.find_element_by_xpath('''//*[@id="SIGUNGU_NM0"]''')
si_gun_gu_list = si_gun_gu_raw.find_elements_by_tag_name("option")
```

```python
from tqdm import tqdm

station_name_list = []
station_address_list = []
logo_list = []
gasoline_price_list = []
diesel_price_list = []
self_yn = []
gu_list = []
wash_yn = []
charger_yn = []
maint_yn = []
cvs_yn = []
sel24_yn = []


for i in tqdm(range(1, len(si_gun_gu_list))):
    si_gun_gu_raw = driver.find_element_by_xpath('''//*[@id="SIGUNGU_NM0"]''')
    si_gun_gu_list = si_gun_gu_raw.find_elements_by_tag_name("option")
    
    driver.find_element_by_xpath('''//*[@id="SIGUNGU_NM0"]''').click() #시군구 클릭
    time.sleep(1)
    si_gun_gu_list[i].click()
    time.sleep(1)
    
    tbody = driver.find_element_by_id("body1")
    station_href_list = tbody.find_elements_by_tag_name("a")
    
    for j in range(0, len(station_href_list)):
        station_href_list[j].click()
        
        req = driver.page_source #현재 페이지의 html코드 가져오기
        soup = BeautifulSoup(req, 'html.parser')
        
        station_info = soup.find(id = "os_dtail_info")
        
        #주유소 이름
        station_name = station_info.find(id="os_nm").get_text()
        station_name_list.append(station_name)
        
        # 주소
        station_address = station_info.find(id="rd_addr").get_text()
        station_address_list.append(station_address)
        
        # 브랜드 로고
        logo = station_info.find(id="os_title_icon")["alt"]
        logo_list.append(logo)
        
        # 휘발유 가격
        gasoline_price = station_info.find(id="b027_p").get_text()
        gasoline_price = gasoline_price.replace(',', '')
        gasoline_price_list.append(gasoline_price)
        
        # 경유 가격
        diesel_price = station_info.find(id="d047_p").get_text()
        diesel_price = diesel_price.replace(',', '')
        diesel_price_list.append(diesel_price)
        
        #셀프
        self = "Y" if station_info.find(id="self_icon") is not None else "N"
        self_yn.append(self)
        
        # 구
        gu = station_address.split(" ")[1]
        gu_list.append(gu)
        
        #세차장
        wash_img = station_info.find(id="cwsh_yn")["src"]
        if charge_img == "/images/user/gis/oil_station_service1_01_off.gif":
            wash_yn.append("N")
        else:
            wash_yn.append("Y")
        
        #충정소
        charge_img = station_info.find(id="lpg_yn")["src"]
        if charge_img == "/images/user/gis/oil_station_service1_02_01_off.gif":
            charger_yn.append("N")
        else:
            charger_yn.append("Y")
            
        #경정비
        maint_yn
        maint_img = station_info.find(id="maint_yn")["src"]
        if maint_img == "/images/user/gis/oil_station_service1_03_off.gif":
            maint_yn.append("N")
        else:
            maint_yn.append("Y")
        
        #편의점
        cvs_yn
        cvs_img = station_info.find(id="cvs_yn")["src"]
        if cvs_img == "/images/user/gis/oil_station_service1_04_off.gif":
            cvs_yn.append("N")
        else:
            cvs_yn.append("Y")
        
        #24시간
        sel24_yn
        sel24_img = station_info.find(id="sel24_yn")["src"]
        if sel24_img == "/images/user/gis/oil_station_service1_05_off.gif":
            sel24_yn.append("N")
        else:
            sel24_yn.append("Y")
        
        
        
```

```python
import pandas as pd

df_oil = pd.DataFrame(
    {
        "station_name":station_name_list,
        "station_address": station_address_list,
        "brand": logo_list,
        "gasoline_price":gasoline_price_list,
        "diesel_price":diesel_price_list,
        "self":self_yn,
        "gu":gu_list,
        "wash":wash_yn,
        "charger":charger_yn,
        "maint":maint_yn,
        "cvs":cvs_yn,
        "sel24":sel24_yn
    }
)

df_oil.head()
```

```python
df_oil["gasoline_price"] = df_oil["gasoline_price"].astype('int64')
df_oil["diesel_price"] = df_oil["diesel_price"].astype('int64')
df_oil.info()
```

### 위도 경도 추가

```python
import googlemaps
import numpy as np
```

```python
gmaps_key = "***********************************"
gmaps = googlemaps.Client(key = gmaps_key)
```

```python
lat = []
lng = []

for idx, row in tqdm(df_oil.iterrows()):

#     if len(row["station_address"].split(')')) < 1:  
#         target_name = row["주소"]
#     else:
#         target_name = ' '.join(row["주소"].split(' ')[:-1])

    target_name = row["station_address"]
    
    gmaps_output = gmaps.geocode(target_name)
    location_output = gmaps_output[0].get("geometry")
    lat.append(location_output["location"]["lat"])
    lng.append(location_output["location"]["lng"])
```

```python
df_oil["lat"] = lat
df_oil["lng"] = lng
df_oil.head()
```

```python
df_oil.to_csv("df_oil.csv", sep=",", encoding="utf-8")
```
