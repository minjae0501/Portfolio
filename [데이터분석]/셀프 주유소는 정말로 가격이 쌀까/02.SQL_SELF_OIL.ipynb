위도, 경도 정보를 이용하여 미왕빌딩에서 셀프주유가 가능하고 24시간이면서 편의점이 있는 가장 가까운 주유소 10개를 휘 발유 가격이 가장 저렴한 순으로 정렬하여 조회하세요.

- 주유소 아이디, 주유소 브랜드명, 주유소 매장명, 주소, 휘발유 가격, 부가정보 (셀프, 24시간, 편의점 여부), 미왕빌딩으로 부터의 거리 (km)


제출 7.
- 쿼리, 결과

```python
sql02 = """SELECT s.id, s.brand, s.name, s.address, s.gasoline, s.self, s.24_hours, s.convenience_store, s.distance
FROM (select *, ST_Distance_Sphere(point(127.0292924,37.4955525), POINT(lng, lat)) AS distance 
 from GAS_STATION
 WHERE self = 1 and 24_hours = 1 and convenience_store = 1
 order by distance
 limit 10) s
order by gasoline
"""

cursor.execute(sql02)
result = cursor.fetchall()
for row in result:
    print(row)
```

문제 8.

구별로 주유소 브랜드 별 휘발유 평균가격을 조회하여 저렴한 순으로 출력하세요.

- 구 이름, 주유소 브랜드 이름, 휘발유 평균 가격

제출 8.
- 쿼리, 결과

```python
sql03 = """SELECT s.gu, b.name, avg(gasoline)
from GAS_STATION s
inner join GAS_BRAND b ON s.brand = b.id
group by s.gu, s.brand
order by avg(gasoline)
"""

cursor.execute(sql03)
result = cursor.fetchall()
for row in result:
    print(row)
```
